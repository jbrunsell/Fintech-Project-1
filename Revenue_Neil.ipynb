```python
# Data from CSV file
df = pd.read_csv('/Users/neilbaride/Downloads/clean_hotel_data.csv')

data = {
    'arrival_date_year': [2015, 2015, 2016, 2016, 2017, 2017],
    'arrival_date_month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'hotel': ['Resort', 'City Hotel', 'City Hotel', 'Resort', 'Resort', 'City Hotel']
}

#  DataFrame
df = pd.DataFrame(data)

# Grouping by year and calculating the count of booking entries per year
yearly_count = df.groupby('arrival_date_year').count()

# Average yearly booking count
average_yearly_count = yearly_count.mean()

# Grouping by month and calculating the count of booking entries per month
monthly_count = df.groupby('arrival_date_month').count()

# Yearly count and he monthly count, and the average yearly count
print("Yearly Booking Count:")
print(yearly_count)
print("\nMonthly Booking Count:")
print(monthly_count)
print(f"\nAverage Yearly Booking Count: {average_yearly_count}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import holoviews as hv

# Data from CSV file
df = pd.read_csv('/Users/neilbaride/Downloads/clean_hotel_data.csv')

# Filtering the data for the year 2015 and 2016
gdp_2015 = df[df['arrival_date_year'] == 2015]['GDP'].sum()
gdp_2016 = df[df['arrival_date_year'] == 2016]['GDP'].sum()

# Bar plot to compare GDP values for 2015 and 2016
years = [2015, 2016]
gdp_values = [gdp_2015, gdp_2016]

plt.figure(figsize=(8, 6))
plt.bar(years, gdp_values, color=['blue', 'green'])
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('Comparison of GDP in 2015 and 2016')
plt.xticks(years)
plt.show()

# Comparing the GDP values for 2015 and 2016
if gdp_2015 > gdp_2016:
    print("GDP in 2015 was higher than 2016.")
elif gdp_2016 > gdp_2015:
    print("GDP in 2016 was higher than 2015.")
else:
    print("GDP in 2015 and 2016 were the same.")
```

```python
import pandas as pd
import hvplot.pandas

# Data from CSV file
df = pd.read_csv('/Users/neilbaride/Downloads/clean_hotel_data.csv')

#  DataFrame is named 'arrival_date_month' 

df['arrival_date_month'] = pd.to_datetime(df['arrival_date_month'], format='%B').dt.month

# Grouping the data by 'arrival_date_year' and 'arrival_date_month' and sum the 'revenue' for each month
monthly_revenue = df.groupby(['arrival_date_year', 'arrival_date_month'])['revenue'].sum().reset_index()

# Bar plot for yearly revenue by month
monthly_revenue.hvplot.bar(
    x='arrival_date_month',
    y='revenue',
    by='arrival_date_year',
    title='Yearly Revenue by Month',
    xlabel='Month',
    ylabel='Revenue',
    yformatter='%.0f' 
)
```

```python
import pandas as pd
import hvplot.pandas

# Data from CSV file
df = pd.read_csv('/Users/neilbaride/Downloads/clean_hotel_data.csv')

# Grouping the data by year and sum the revenue for each year
yearly_revenue = df.groupby(df['arrival_date_year'])['revenue'].sum().reset_index()

# Creating a new column 'quarter' to represent the quarter
yearly_revenue['quarter'] = pd.to_datetime(yearly_revenue['arrival_date_year'], format='%Y').dt.to_period('Q')

# Groupping the data by 'quarter' and sum the 'revenue' for each quarter
quarterly_revenue = yearly_revenue.groupby('quarter')['revenue'].sum().reset_index()

# Bar plot for revenue by quarter with whole number y-axis labels
quarterly_revenue_plot = quarterly_revenue.hvplot.bar(
    x='quarter',
    y='revenue',
    title='Revenue by Quarter',
    yformatter="%.0f"  
)

# Show the plot 
quarterly_revenue_plot
```
